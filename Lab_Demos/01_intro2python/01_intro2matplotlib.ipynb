## A quick visualization of data:

```python
import numpy as np 
# this is the main plotting library in python
# get use to importing this 
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2,0.1)

# Plot some data on the axes.
plt.plot(x,x,   label='linear')
# Plot more data on the axes...
plt.plot(x,x**2,label='quadratic')
# ... and some more.
plt.plot(x,x**3,label='cubic')     
# Add an x-label t
plt.xlabel('x label')  
# Add a y-label 
plt.ylabel('y label')  
# Add a title 
plt.title("Simple Plot")  
# Add a legend.
plt.legend()  
```

## Creating figure and axes objects

```python
x = np.arange(0,2,0.1)


# Create a figure and an axes.
fig1, ax = plt.subplots(nrows=1,ncols=1)

# Plot some data on the axes.
ax.plot(x,x,   label='linear')   

# Plot more data on the axes...
ax.plot(x,x**2,label='quadratic')

# ... and some more.
ax.plot(x,x**3,label='cubic')    
# Add an x-label to the axes.
ax.set_xlabel('x label')  
# Add a y-label to the axes.
ax.set_ylabel('y label')  
# Add a title to the axes.
ax.set_title("Simple Plot")  
# Add a legend.
ax.legend()  

# Prevent plot from displaying in this cell
plt.close()
```

## Creating figure and multiple axes objects

```python
# Create a figure and two axes.
fig2, ax = plt.subplots(nrows=2,ncols=1)
```

```python
ax
```

```python
# Create an x array
x = np.arange(0,2*np.pi,0.1)

# Create a figure and two axes with a shared x-axis
fig2, ax = plt.subplots(nrows=2,ncols=1, sharex=True)
# Plot some data on the first axes.
ax[0].plot(x,np.sin(x),label='sin(x)') 
# Plot some data on the second axes.
ax[1].plot(x,np.cos(x),label='cos(x)')
# Add a y-label to the axes one
ax[0].set_ylabel('Amplitude')  
# Add a y-label to the axes two
ax[1].set_ylabel('Amplitude')  
# Add a x-label to the axes two
ax[1].set_xlabel("x-axis")  
# Add a title to the axes one.
ax[0].set_title("Simple Plot")  
# Add a legend to axes one.
ax[0].legend() 
# Add a legend to axes two.
ax[1].legend()
#make the plot look nice 
fig2.tight_layout()

# Prevent plot from displaying in this cell
plt.close()
```
